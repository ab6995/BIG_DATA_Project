```python
bronx_assault = pd.read_csv('/Users/ashishbansal/Downloads/ds1004-big-data-final-project-master/Datasets/timeresult/bronx-assault.csv', names = ['Hour','Frequency'])
bronx_harassment = pd.read_csv('/Users/ashishbansal/Downloads/ds1004-big-data-final-project-master/Datasets/timeresult/bronx-harrasment.csv', names = ['Hour','Frequency'])
bronx_larceny = pd.read_csv('/Users/ashishbansal/Downloads/ds1004-big-data-final-project-master/Datasets/timeresult/bronx-larceny.csv', names = ['Hour','Frequency'])
```

```python
p1 = figure(title=" Bronx - Top 3 Reported Incidents by the Hour of the Day")
p1.grid.grid_line_alpha=0.3
p1.xaxis.axis_label = 'Hour'
p1.yaxis.axis_label = 'Frequency of Occurrence'
#p1.xaxis[0].ticker.desired_num_ticks = 24 #adjust the number of ticks for time axis

p1.line(bronx_assault['Hour'], bronx_assault['Frequency'], color='#B2DF8A',line_width = 2, legend = 'Assault')
p1.line(bronx_harassment['Hour'], bronx_harassment['Frequency'], color='#fc8d59',line_width = 2, legend = 'Harrassment')
p1.line(bronx_larceny['Hour'], bronx_larceny['Frequency'], color='#21618C',line_width = 2, legend = 'Larceny')

show(p1)
```
